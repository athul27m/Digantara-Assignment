```python
df = pd.read_csv('sort-minRange.csv')
```

```python
#CONVERT TCA to datetime

df['TCA'] = pd.to_datetime(df['TCA'])
df.info()
```

```python
#Check for null values

df.isnull().sum()
```

```python
df.describe().T
```

```python
#Check for Duplicates

df.duplicated().value_counts()
```

Descriptive Analytics


```python
#Total Conjunctions

conjunctions = len(df)
print(f"Total Number of Conjuntions Active :{conjunctions}")
```

```python
#Active Satellites

active_satellites = df['NORAD_CAT_ID_1']._append(df['NORAD_CAT_ID_2']).unique()
print(f"Total Number of Active Satellites:{len(active_satellites)}")
```

```python
# Risk Management

low_risk = df[df['MAX_PROB'] < 0.05]
high_risk = df[df['MAX_PROB'] >= 0.05]

print(f"Low Risk Conjunctions: {len(low_risk)}")
print(f"High Risk Conjunctions: {len(high_risk)}")
```
